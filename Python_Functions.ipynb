### Additional Function Codes

```python
def is_even(num):
  return num % 2 == 0

print(f"Is 4 even? {is_even(4)}")
print(f"Is 7 even? {is_even(7)}")
```

```python
def is_odd(num):
  return num % 2 != 0

print(f"Is 4 odd? {is_odd(4)}")
print(f"Is 7 odd? {is_odd(7)}")
```

```python
def factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    result = 1
    for i in range(2, n + 1):
      result *= i
    return result

print(f"Factorial of 5: {factorial(5)}")
print(f"Factorial of 0: {factorial(0)}")
```

```python
def fibonacci(n):
  a, b = 0, 1
  series = []
  while a <= n:
    series.append(a)
    a, b = b, a + b
  return series

print(f"Fibonacci series up to 10: {fibonacci(10)}")
```

```python
def is_prime(num):
  if num < 2:
    return False
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False
  return True

print(f"Is 11 prime? {is_prime(11)}")
print(f"Is 10 prime? {is_prime(10)}")
```

```python
def calculate_power(base, exp):
  return base ** exp

print(f"2 to the power of 3: {calculate_power(2, 3)}")
```

```python
import math
def square_root(num):
  if num < 0:
    return "Cannot calculate square root of a negative number"
  return math.sqrt(num)

print(f"Square root of 25: {square_root(25)}")
```

```python
def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32

print(f"25 Celsius to Fahrenheit: {celsius_to_fahrenheit(25)}")
```

```python
def fahrenheit_to_celsius(fahrenheit):
  return (fahrenheit - 32) * 5/9

print(f"77 Fahrenheit to Celsius: {fahrenheit_to_celsius(77)}")
```

```python
def reverse_string(s):
  return s[::-1]

print(f"Reversed 'hello': {reverse_string('hello')}")
```

```python
def is_palindrome(s):
  cleaned_s = "".join(filter(lambda char: char.isalnum(), s)).lower()
  return cleaned_s == cleaned_s[::-1]

print(f"Is 'madam' a palindrome? {is_palindrome('madam')}")
print(f"Is 'A man, a plan, a canal: Panama' a palindrome? {is_palindrome('A man, a plan, a canal: Panama')}")
```

```python
def count_vowels(s):
  vowels = "aeiouAEIOU"
  count = 0
  for char in s:
    if char in vowels:
      count += 1
  return count

print(f"Number of vowels in 'Education': {count_vowels('Education')}")
```

```python
def remove_spaces(s):
  return s.replace(" ", "")

print(f"'Hello World' without spaces: {remove_spaces('Hello World')}")
```

```python
def find_max(numbers):
  if not numbers:
    return None
  return max(numbers)

print(f"Max of [1, 5, 2, 8]: {find_max([1, 5, 2, 8])}")
```

```python
def find_min(numbers):
  if not numbers:
    return None
  return min(numbers)

print(f"Min of [1, 5, 2, 8]: {find_min([1, 5, 2, 8])}")
```

```python
def calculate_sum(numbers):
  return sum(numbers)

print(f"Sum of [1, 2, 3, 4]: {calculate_sum([1, 2, 3, 4])}")
```

```python
def calculate_average(numbers):
  if not numbers:
    return 0
  return sum(numbers) / len(numbers)

print(f"Average of [1, 2, 3, 4]: {calculate_average([1, 2, 3, 4])}")
```

```python
def sort_list_asc(numbers):
  return sorted(numbers)

print(f"Sorted ascending [5, 2, 8, 1]: {sort_list_asc([5, 2, 8, 1])}")
```

```python
def sort_list_desc(numbers):
  return sorted(numbers, reverse=True)

print(f"Sorted descending [5, 2, 8, 1]: {sort_list_desc([5, 2, 8, 1])}")
```

```python
def remove_duplicates(items):
  return list(set(items))

print(f"Remove duplicates from [1, 2, 2, 3, 1]: {remove_duplicates([1, 2, 2, 3, 1])}")
```

```python
def greet_user(name):
  return f"Hello, {name}! Welcome."

print(greet_user("Alice"))
```

```python
def area_of_circle(radius):
  import math
  return math.pi * (radius ** 2)

print(f"Area of circle with radius 5: {area_of_circle(5)}")
```

```python
def volume_of_cube(side):
  return side ** 3

print(f"Volume of cube with side 3: {volume_of_cube(3)}")
```

```python
def simple_interest(principal, rate, time):
  return (principal * rate * time) / 100

print(f"Simple interest for P=1000, R=5, T=2: {simple_interest(1000, 5, 2)}")
```

```python
def calculate_bmi(weight_kg, height_m):
  if height_m <= 0:
    return "Height must be positive"
  return weight_kg / (height_m ** 2)

print(f"BMI for 70kg, 1.75m: {calculate_bmi(70, 1.75):.2f}")
```
