[1,2,3,4,5] → These are your x-values.

[1,4,9,16,25] → These are your y-values (squares of x-values).

'r^-' → This is a format string that styles the plot:

'r' → Red color

'^' → Triangle-up markers for each point

'-' → Solid line connecting the points

```python
plt.grid(True)
plt.title('squares of the numbers')
plt.xlabel('The Numbers', fontsize=18, color='red')
plt.ylabel('The squares',fontsize=18, color='red')
plt.plot([1,2,3,4,5], [1,4,9,16,25], 'r^--', linewidth=3.0)
plt.show()
```

### 3D Scatter Plot Example

This example demonstrates how to create a basic 3D scatter plot using `matplotlib`'s `mplot3d` toolkit.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Sample data for a 3D scatter plot
np.random.seed(42)
x_data = np.random.rand(100)
y_data = np.random.rand(100)
z_data = np.random.rand(100)

# Create a new figure and add a 3D subplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the scatter points
ax.scatter(x_data, y_data, z_data, c='blue', marker='o')

# Set labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set a title for the plot
ax.set_title('Simple 3D Scatter Plot')

# Show the plot
plt.show()
```

### 3D Line Plot Example

This example creates a 3D line plot, often used to visualize trajectories or sequences in three dimensions.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Sample data for a 3D line plot (a helix)
t = np.linspace(-2 * np.pi, 2 * np.pi, 500)
x_line = np.sin(t)
y_line = np.cos(t)
z_line = t

# Create a new figure and add a 3D subplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D line
ax.plot(x_line, y_line, z_line, color='red', linewidth=2)

# Set labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set a title for the plot
ax.set_title('Simple 3D Line Plot (Helix)')

# Show the plot
plt.show()
```
