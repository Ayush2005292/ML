```python
List = [[2,3,4],[1,24,16,64],[3,6,9,12]]

sortList = lambda x: [sorted(i) for i in x]
print(sortList(List))

secondlargest = lambda x:[y[-2] for y in sortList(x)]
print(secondlargest(List))
```

```python
from functools import reduce
li = [5,8,10,20,50,100]
sum = reduce((lambda x,y: x+y), li)
print(sum)
```
