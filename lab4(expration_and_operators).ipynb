write a program in python to evaluate 1! + 2! + 3! +  ..........n!

```python
import math

def factorial_sum(n):
    sum = 0
    for i in range(1, n + 1):
        sum += math.factorial(i)
    return sum

num = int(input("Enter a number (n) to calculate the sum of factorials up to n!: "))

result = factorial_sum(num)
print(f"The sum of factorials from 1! to {num}! is: {result}")
```

wap in python to enter the number till the user wants and display the maximum number oot of those number

```python
max_num = 0

while True:
  num = int(input("Enter number (a to stop): "))
  if num == 0:
    break
    if num>max_num:
      max_num = num
    print(max_num)
```

```python
class Car:
  def __init__(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year

  def display_info(self):
    print(f'Car: {self.make} {self.model} {self.year}')

# Example usage:
my_car = Car("Toyota", "Camry", 2022)
my_car.display_info()
```
